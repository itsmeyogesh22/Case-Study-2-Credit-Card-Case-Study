## a. In case age is less than 18, replace it with mean of age values.

### Step-1: Get all customers that belong to  Age < 18 group

```python
maskFilter1 = customer_acquisition['Age'] < 18
customer_acquisition[maskFilter1 == True]\
                                         .style\
                                         .format({"Limit" : "{:,.2f}"})\
                                         .background_gradient(cmap = 'Blues', subset = 'Age')
```

```python
indexFilter = [13, 40, 62, 67, 69, 89]
customer_acquisition['Age'] = customer_acquisition['Age'].apply(lambda x: customer_acquisition['Age']\
                                                         .mean() if x < 18 else x)

customer_acquisition.loc[indexFilter, :]\
                                        .style\
                                        .background_gradient(cmap = 'Blues', subset = 'Age')\
                                        .format({"Limit" : "{:,.2f}", 
                                                 "Age" : "{:.0f}"})
```

```python
customer_acquisition['Age'] = customer_acquisition['Age'].astype('object')
customer_acquisition['Age'] = customer_acquisition['Age'].astype('int')
```

## b. In case spend amount is more than the limit, replace it with 50% of that customer’s limit. (customer’s limit provided in acquisition table is the per transaction limit on his card)

### Step-1: Merge the datasets to get suitable columns

```python
customer_spends = pd.merge(customer_acquisition, 
                           spend, 
                           on = 'Customer', 
                           how = 'inner')
```

### Step-2: Get 50% of the spending limit for each Customer

```python
customer_spends['50% of limit'] = customer_spends['Limit']/100*50
```

### Step-3: Determine which transactions spent more than the limit

```python
def spendLimit(Amount, Limit):
    spend_amt = Limit/100*50 
    if Amount > Limit:
        return spend_amt
    else:
        return Amount
```

### Step-4: Get the Customers spending more than the limit

```python
limitFilter1 = customer_spends['Limit'] < customer_spends['Amount']
print("\033[1mTotal number of unique customers who made transactions exceeding the card limit: \033[0m",
      customer_spends.loc[limitFilter1, 'Customer'].nunique())
```

```python
print("\033[1m Count of transactions made that exceed limit by each customer: \033[0m")
custMetric1 = pd.DataFrame(customer_spends[limitFilter1].groupby(['Customer'])['Amount']\
                                                        .agg('count'))\
                                                        .rename(columns = {'Amount' : 'transaction_count'})\
                                                        .sort_values('transaction_count')

custMetric1.style.background_gradient(cmap = 'Blues', subset = 'transaction_count')
```

### Step-5: Get column with replacement values

```python
customer_spends['spend_amount'] = np.array(zip(customer_spends['Amount'], customer_spends['Limit']))
customer_spends['spend_amount'] = customer_spends['spend_amount'].apply(lambda x: spendLimit(x[0], x[1]))
customer_spends['Exceeded Limit'] = customer_spends['Amount'] > customer_spends['Limit']
customer_spends.loc[limitFilter1, ['Customer', 'Limit', 'Amount', '50% of limit', 'spend_amount']]\
               .sample(20)\
               .style.background_gradient(cmap = 'Blues', subset = ['Amount', 'spend_amount'])\
               .format({"Amount" : "{:,.2f}", "spend_amount" : "{:,.2f}", "50% of limit" : "{:,.2f}", "Limit" : "{:,.2f}"})
```

### Step-6: Replace values in the actual 'spend' dataset (inplace)

```python
spend = pd.merge(spend[['Sl No:', 'Customer', 'Month', 'Type', 'Amount']], 
                 customer_spends[['Customer', 'Month', 'Limit', 'spend_amount', 'Exceeded Limit']], 
                 on = ['Customer', 'Month'],
                 how = 'inner'
                )
spend.drop(labels = ['Amount', 'Limit', 'Exceeded Limit'], axis = 'columns', inplace = True)
spend.rename(columns = {"spend_amount" : "Amount"}, inplace = True)
spend.sample(15)
```

### Step-7: Verify the results | Check if there are any transactions left that exceed the limit

```python
check = pd.merge(spend[['Sl No:', 'Customer', 'Month', 'Type', 'Amount']], 
                 customer_spends[['Customer', 'Month', 'Limit', 'spend_amount', 'Exceeded Limit']], 
                 on = ['Customer', 'Month'], 
                 how = 'inner')

check['Exceeded Limit'] = check['Amount'] > check['Limit']
check['Exceeded Limit'] = check['Exceeded Limit'].apply(lambda x: 'Yes' if x == True else 'No')
check.sample(15)
```

```python
print("\033[1mTransactions that exceeded limit post data updation: \033[0m", 
      check.loc[check['Exceeded Limit'] == 'Yes', ['Customer']]['Customer'].count())
```

```python
custMetrics2 = pd.DataFrame(customer_pays.loc[customer_pays['Exceeded Limit'] == True, ['Customer', 'Exceeded Limit']]\
                                         .groupby(['Customer'])['Exceeded Limit']\
                                         .agg('count'))

custMetrics2.rename(columns = {'Exceeded Limit' : 'repayment_account'}, inplace = True)
custMetrics2.style.background_gradient(cmap = 'Blues', subset = 'repayment_account')
```

### Step-3: Determine which repayment account exceeds more than the limit

```python
def spendLimit(Amount, Limit):
    if Amount > Limit:
        return Limit
    else:
        return Amount
```

### Step-4: Get column with replacement values

```python
customer_pays['revised_repayment'] = np.array(zip(customer_pays['Amount'], customer_pays['Limit']))
customer_pays['revised_repayment'] = customer_pays['revised_repayment'].apply(lambda x: spendLimit(x[0], x[1]))
customer_pays['Exceeded Limit'] = customer_pays['Amount'] > customer_pays['Limit']
customer_pays.loc[:, ['Customer', 'Product', 'Amount', 'Limit', 'revised_repayment']]\
             .sample(20)\
             .style\
             .background_gradient(cmap = 'Blues', subset = ['Amount', 'revised_repayment'])\
             .format({"Amount" : "{:,.2f}", "Limit" : "{:,.2f}", "revised_repayment" : "{:,.2f}"})
```

### Step-5: Replace values in the actual 'spend' dataset (inplace)

```python
repayment = pd.merge(repayment[['SL No:', 'Customer', 'Month', 'Amount']], 
                     customer_pays[['Customer', 'Month', 'Limit', 'revised_repayment']], 
                     on = ['Customer', 'Month'], 
                     how = 'inner')
repayment.drop(labels = ['Amount', 'Limit'], axis = 'columns', inplace = True)
repayment.rename(columns = {"revised_repayment" : "Amount"}, inplace = True)
repayment.sample(15)
```

### Step-6: Verify the results | Check if there are any repayment accounts left that exceed the limit

```python
check = pd.merge(repayment[['SL No:', 'Customer', 'Month', 'Amount']], 
                 customer_pays[['Customer', 'Month', 'Limit', 'Exceeded Limit']], 
                 on = ['Customer', 'Month'], 
                 how = 'inner')

check['Exceeded Limit'] = check['Amount'] > check['Limit']
check['Exceeded Limit'] = check['Exceeded Limit'].apply(lambda x: 'Yes' if x == True else 'No')
check.sample(15)\
     .style\
     .background_gradient(cmap = 'Blues', subset = ['Amount', 'Limit'])\
     .format({"Amount" : "{:,.2f}", "Limit" : "{:,.2f}"})
```

## e. If the monthly rate of interest is 2.9%, what is the profit for the bank for each month?(Profit is defined as interest earned on Monthly Profit. Monthly Profit = Monthly repayment – Monthly spend. Interest is earned only on positive profits and not on negative amounts)

```python
AvgProfitStyler = [dict(selector = 'caption', props = [('font-weight', 'bold'), 
                                                       ('font-size', '150%'), 
                                                       ('text-align', 'center')])]

profit_metric_group1 = pd.merge(spendMetric1, 
                                repayMetric1, 
                                on = ['Year', 'Month'], 
                                how = 'inner', 
                                suffixes = [' Spent', ' Repayed'])

profit_metric_group1['Monthly Profit'] = profit_metric_group1['Amount Repayed'] - profit_metric_group1['Amount Spent']
profit_metric_group1['Profit'] = np.where(profit_metric_group1['Monthly Profit'] > 0, 
                                          (profit_metric_group1['Monthly Profit']*2.9)/100, 0)
profit_metric_group1['Profit Margin'] = (profit_metric_group1['Monthly Profit']/profit_metric_group1['Amount Repayed'])*100

profit_metric_group1.style\
                    .background_gradient(cmap = 'YlGnBu_r', subset = ['Amount Spent', 
                                                                      'Amount Repayed', 
                                                                      'Monthly Profit', 
                                                                      'Profit', 
                                                                      'Profit Margin'])\
                    .highlight_between(left = -100000.000, right = 0.0, color = 'crimson', subset = ['Monthly Profit', 
                                                                                                     'Profit', 
                                                                                                     'Profit Margin'])\
                    .set_caption("Average Monthly Profit<br><br>").set_table_styles(AvgProfitStyler)\
                    .format({"Amount Spent" : "{:,.2f}", 
                             "Amount Repayed" : "{:,.2f}", 
                             "Monthly Profit" : "{:,.2f}", 
                             "Profit" : "{:,.2f}"})
```

### Top-10 Customers by specific Product and time period (specific Year -OR- Month)

```python
customer_pays1 = pd.merge(customer_acquisition, 
                         repayment, 
                         on = 'Customer', 
                         how = 'inner')

repayment_group2 = customer_pays1[['City', 'Product', 'Month', 'Customer', 'Amount']]
repayment_group2['Year'] = repayment_group2['Month'].dt.year
repayment_group2['Month Name'] = repayment_group2['Month'].dt.month_name()
repayment_group2.drop('Month', axis = 'columns', inplace = True)
repayment_group2 = repayment_group2[['Year', 'Month Name', 'City', 'Product', 'Customer', 'Amount']]


def repaymentMetrics1(period, productType):
    if type(period) == str:
        result_set_month = pd.DataFrame(repayment_group2.groupby(['Month Name', 
                                                                  'City', 
                                                                  'Customer', 
                                                                  'Product'])['Amount']\
                                        .agg('sum')).reset_index()
        
        result_set_month = result_set_month.loc[(result_set_month['Month Name']==period)\
                                               &(result_set_month['Product']==productType),:]\
                                          .sort_values('Amount', ascending = False)\
                                          .groupby('City').head(10)
        
        result_set_month = result_set_month.sort_values('City', ascending = True)
        result_set_month = result_set_month.pivot_table(index = ['Month Name', 
                                                                 'City', 
                                                                 'Customer', 
                                                                 'Product'], 
                                                        values = 'Amount')
        return result_set_month
    elif type(period) == int:
        result_set_year = pd.DataFrame(repayment_group2.groupby(['Year', 
                                                                 'City', 
                                                                 'Customer', 
                                                                 'Product'])['Amount'].agg('sum')).reset_index()
        
        result_set_year = result_set_year.loc[(result_set_year['Year']==period)\
                                              &(result_set_year['Product']==productType),:]\
                                              .sort_values('Amount', ascending = False)\
                                              .groupby('City').head(10)
        
        result_set_year = result_set_year.sort_values('City', ascending = True)
        result_set_year = result_set_year.pivot_table(index = ['Year', 'City', 'Customer', 'Product'], values = 'Amount')
        return result_set_year
    
repaymentMetrics1('January', 'Gold')
```
